```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv('/content/winequality.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['fixed acidity']= df['fixed acidity'].fillna(df['fixed acidity'].median())
df.isnull().sum()
```

```python

q1=df['quality'].quantile(0.25)
q3=df['quality'].quantile(0.75)
iqr=q3-q1
upper=q3+q3*iqr
df=df[(df['quality']<=upper)]
df.shape
```

```python
x = df.drop(columns="quality")
y = df["quality"]
```

```python
print(x)
```

```python
print(y)
```

```python
scaler = StandardScaler()
x_std = scaler.fit_transform(x)
```

```python
print(x_std)
```

```python
x=x_std
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
cf = RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    random_state=2
)
```

```python
cf.fit(x_train, y_train)
```

```python

x_train_pred=cf.predict(x_train)
```

```python
x_train_r2 = r2_score(y_train, x_train_pred)
```

```python
print("Training R2 Score:", x_train_r2)
```

```python
x_test_pred=cf.predict(x_test)
x_test_r2=r2_score(y_test,x_test_pred)
print(x_test_r2)
```
